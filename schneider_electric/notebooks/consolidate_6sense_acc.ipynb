```python
import pandas as pd
import os
```

```python
def read_analytics_data(file_path):
    xls = pd.ExcelFile(file_path)
    possible_sheets = ["Analytics By Accounts", "Analytics By Account", "Analytics By Accouts"]
    
    for sheet in possible_sheets:
        if sheet in xls.sheet_names:
            sheet_name = sheet
            
    # Read the sheet with headers in row 7 and data starting from row 8
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=6)

    # Read column names from A1, A2, A3, A4 and their values from B1, B2, B3, B4
    extra_values = pd.read_excel(file_path, sheet_name=sheet_name, usecols="A,B", nrows=4, header=None)
    column_names = extra_values.iloc[:, 0].values  # Get names from column A
    column_values = extra_values.iloc[:, 1].values  # Get values from column B

    # Assign values to new columns dynamically
    for col_name, col_value in zip(column_names, column_values):
        df[col_name] = col_value  # Set the same value for all rows

    df["file name"] = os.path.splitext(os.path.basename(file_path))[0]

    return df
```

```python
df = read_analytics_data("/home/darshan/Documents/Schneider Electric Data/6sense Ads/Robotics 6sense Ads/SE_Robotics_18_Nov_2024.xlsx")
print(df.head())
df.info()
```

```python
folder_path = "/home/darshan/Documents/Schneider Electric Data/6sense Ads"
all_data = []

for root, _, files in os.walk(folder_path):
    for file in files:
        print(file)
        if file.endswith(".xlsx"):
            file_path = os.path.join(root, file)
            df = read_analytics_data(file_path)
            print(df.shape)
            print(df.head(1))
            all_data.append(df)

final_df = pd.concat(all_data, ignore_index=True)
```
