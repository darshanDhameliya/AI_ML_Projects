```python
# Precompile the regex for efficiency
pattern = re.compile(r'(?<=_)([^_]+)(?=_)')
```

```python
def match_segment(extracted_segment):
    """Efficiently finds the best match for extracted_segment in predefined segments."""
    extracted_segment = str(extracted_segment).lower().replace(" ","")  # Ensure string and lowercase
    for segment_lower in segments_lower:
        if segment_lower.replace(" ","") in extracted_segment or extracted_segment in segment_lower.replace(" ",""):
            return segment_lower  # Return original formatted segment
    return None
```

```python
def extract_segment_vectorized(series):
    """Vectorized function to extract & map segments in Pandas."""
    extracted = series.str.extract(pattern)[0].str.lower()  # Extract first match & lowercase
    return extracted.map(match_segment)  # Apply match function
```

```python
unique_prospect_engagements["segment"] = extract_segment_vectorized(unique_prospect_engagements["source"])
```
